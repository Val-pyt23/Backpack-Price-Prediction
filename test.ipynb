## Brand Imputation

```python
brand_known = data_train[data_train['Brand'].notnull()].copy()
brand_known_missing = data_train[data_train['Brand'].isnull()].copy()

features = ['Material', 'Size', 'Compartments', 'Laptop Compartment',
            'Waterproof', 'Style', 'Color', 'Weight Capacity (kg)']
```

```python
# impute
for col in features:
    if data_train[col].dtype == 'object':
        mode = data_train[col].mode()[0]
        brand_known[col].fillna(mode,inplace=True)
        brand_known_missing[col].fillna(mode,inplace=True)
    else:
        median = data_train[col].median()
        brand_known[col].fillna(median,inplace=True)
        brand_known_missing[col].fillna(median,inplace=True)

data_train_combined = pd.concat([brand_known,brand_known_missing])
for col in features:
    if data_train_combined[col].dtype == 'object':
        le = LabelEncoder()
        data_train_combined[col] = le.fit_transform(data_train_combined[col])
```

```python
# pisah buat prediksi
x_brand = data_train_combined.loc[brand_known.index, features]
x_brand_missing = data_train_combined.loc[brand_known_missing.index, features]

brand_known = brand_known[brand_known['Brand'].notnull()]
le_brand = LabelEncoder()
y_brand = le_brand.fit_transform(brand_known['Brand'])

rforest = RandomForestClassifier(n_estimators=100,random_state=42)
rforest.fit(x_brand,y_brand)

pred_encoded = rforest.predict(x_brand_missing)
pred_labels = le_brand.inverse_transform(pred_encoded)

data_train.loc[brand_known_missing, 'Brand'] = pred_labels

print("Sisa Brand yang missing:", data_train['Brand'].isnull().sum())
```
